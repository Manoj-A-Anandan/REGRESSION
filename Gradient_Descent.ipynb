```python
def gradient_descent(x,y):
  m_curr=b_curr=0
  iterations=1000
  alpha=0.08
  n=len(x)
  y_p=[]
  for i in range(iterations):
    y_pred=b_curr+m_curr*x
    y_p.append(y_pred)
    cost=(1/n)*sum([val**2 for val in(y-y_p)])
    md=-(2/n)*sum(x*(y-y_pred))
    bd=-(2/n)*sum(y-y_pred)
    b_curr=b_curr-alpha*bd
    m_curr=m_curr-alpha*md
    print('M {} C {} cost {} Ieration {}'.format(m_curr,b_curr,cost,i))
  return y_p
import matplotlib.pyplot as plt
import numpy as np
x=np.array([1,2,3,4,5])
y=np.array([5,7,9,11,13])
y_p=gradient_descent(x,y)
plt.scatter(x,y)
for i in y_p:
  plt.plot(x,i)
plt.xlabel('x data')
plt.ylabel('y data')
plt.title('717822i232 - MANOJ A')
```
